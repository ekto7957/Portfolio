````python
# 섹션 2 채우기 함수  + 다른응답버튼 포함
def fill_form(name_,company_,position_):
   
    driver.find_element(By.XPATH,'//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div/span').click()
    inputs = driver.find_elements(By.CSS_SELECTOR,'[type="text"]')
       
    inputs_arr = [
        name_,company_,position_
        ]
    for i in range(len(inputs)):
        inputs[i].clear()
        inputs[i].send_keys(inputs_arr[i])

    # 2섹션 제출버튼 
    submit = driver.find_element(By.XPATH,'//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div[2]/span')
    submit.click()
   
    another_response = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div/div[4]/a')
    another_response.click()
    
````

````python
# 입력한 데이터 모두 지우기
for i in range(len(inputs)):
    inputs[i].clear()
````

````python
df2 = df[['성명','소속(기업명)','직책']]
len(df2)
````

````python
for i in range(len(df2)):

    fill_form(df2['성명'][i],df2['소속(기업명)'][i],df2['직책'][i])
````
